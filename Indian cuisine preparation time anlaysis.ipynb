```python
p=np.arange(len(df))
width=0.5
p1=[z+width for z in i]
plt.xlabel("Dishes",fontsize=20)
plt.ylabel("Time in minute",fontsize=20)
plt.title("Cooking Time vs Preparation Time",fontsize=20)
width=0.3
plt.bar(p,df['Cooking Time (minutes)'],width,color="b",label="Cooking Time",edgecolor='k')
plt.bar(p1,df['Preparation Time (minutes)'],width,color="#FF0000",label="Preparation Time",edgecolor='k')
plt.xticks(p+width/2,df['Dish'],rotation=90)
plt.legend()
plt.show()
```

The graph titled **"Cooking Time vs Preparation Time"** compares the cooking and preparation times for various dishes. Here are some insights:

- **Cooking Time Dominates**: For most dishes, the cooking time (blue bars) is longer than the preparation time (red bars).
- **Preparation-Heavy Dishes**: A few dishes have preparation times that are equal to or exceed the cooking times, indicating they require more effort in chopping, marinating, or assembling ingredients.
- **Varied Times**: Both cooking and preparation times vary significantly across different dishes, highlighting the diversity in cooking processes.

This visual representation helps in understanding which dishes are more time-consuming in terms of preparation versus cooking.
